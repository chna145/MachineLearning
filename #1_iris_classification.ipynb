```python
# 붓꽃 데이터셋 불러오기
from sklearn.datasets import load_iris
iris_dataset = load_iris()
```

```python
# 붓꽃 데이터셋의 키
print("붓꽃 데이터셋의 키: ", iris_dataset.keys())
```

```python
# 훈련 데이터와 테스트 데이터 나누기
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
        iris_dataset['data'],iris_dataset['target'], test_size=0.3, random_state=10, shuffle=True, stratify=iris_dataset.target)
```

```python
# 훈련 데이터를 데이터프레임으로 변환
import pandas as pd
iris_df = pd.DataFrame(X_train,columns=iris_dataset.feature_names)
```

```python
# 훈련 데이터의 산점도 행렬
pd.plotting.scatter_matrix(iris_df, c=y_train, figsize=(20,20), marker='o',#grid=True,
                           s=50, alpha=1.0)
```

```python
pd.plotting.scatter_matrix(iris_df, c=y_train, figsize=(20,20), marker='o',grid=False,
                           s=50, alpha=1.0)
```

```python
# K-NN 모델 만들기
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
# K-NN 모델 학습시키기
knn.fit(X_train,y_train)
```

```python
# K-NN 모델로 예측하기
y_pre = knn.predict(X_test)
```

```python
# K-NN 모델 정확도
knn.score(X_test,y_test)
```
